### Important Attributes

```python

pd.read_csv('Iris.csv',
            sep=',',
            names=['Id', 'SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm', 'Species'],
            index_col= 'Id',
            skiprows=[1, 2, 3],
            na_values=[np.nan, 'NAN', 'na'],
            nrows=5
)
```

```python
handle_missing_vals = {
    'SepalLengthCm': 'NAN',
    'SepalWidthCm': 'na'
}
pd.read_csv('Iris.csv', na_values=handle_missing_vals).head()
```

```python
iterator = pd.read_csv('Iris.csv', chunksize=100)

for data in iterator:
    print(data)
    break # First 100 rows
```

### CSV Module

```python
import csv
```

```python
f = open('Iris.csv')
reader = csv.reader(f) # Returns a list containing values of each record
```

```python
for line in reader:
    print(line)
```

```python
with open('Iris.csv') as f:
    lines = list(csv.reader(f))

header, values = lines[0], lines[1::]
```

```python
temp = zip(*values)
for i in temp:
    print(i)
```

```python
data_dict = {k: v for k, v in zip(header, zip(*values))} # Consider *values as destructuring the list values
data_dict
```
